## Day-1 --> How to create a dataframe

```
from datetime import date
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType, ArrayType, MapType
```

```
data = [(1, 'karthik', 50, '2020/11/28'), (2, 'Akash', 49, '2020/11/30')]
data_1 = [(1, 'karthik', 50, date(2020, 11, 28)), (2, 'Akash', 49, date(2020, 11, 28))]
row_data = [Row(1, 'karthik', 50, date(2020, 11, 28)), Row(2, 'Akash', 49, date(2020, 11, 28))]

complex_type_data = [
  (2, "Brooke","Wenig", "https://tinyurl.2", "5/5/2018", 8908, ["twitter", "LinkedIn"], {'twitter':2, 'LinkedIn':10}),
  (3, "Denny", "Lee", "https://tinyurl.3", "6/7/2019", 7659, ["web", "twitter", "FB", "LinkedIn"], {'web':2, 'LinkedIn':10}),
  (4, "Tathagata", "Das", "https://tinyurl.4", "5/12/2018", 10568, ["twitter", "FB"], {'FB':2, 'LinkedIn':10})
]

columns = ['Id', 'Name', 'Age', 'DOB']

schema = StructType([
  StructField('Id', IntegerType(), True), 
  StructField('Name', StringType(), True), 
  StructField('Age', IntegerType(), True), 
  StructField('DOB', DateType(), True)
])

schema_complex_type = StructType([
  StructField("Id", IntegerType(), True),
  StructField("First", StringType(), True),
  StructField("Last", StringType(), True),
  StructField("Url", StringType(), True),
  StructField("Published", StringType(), True),
  StructField("Hits", IntegerType(), True),
  StructField("Campaigns", ArrayType(StringType(), True), True),
  StructField("CampaignsCount", MapType(StringType(), IntegerType(), True), True)
])
  
schema_1 = "Id INT, Name STRING, Age INT , dob DATE"

def create_dataframe_basic_datatypes(data):
  df = spark.createDataFrame(data)
  return df

def create_dataframe_basic_datatypes_with_column(data, columns):
  df = spark.createDataFrame(data, columns)
  return df

def create_dataframe_with_schema(data, schema):
  df = spark.createDataFrame(data = data, schema = schema)
  return df
```

```
df_no_cols = create_dataframe_basic_datatypes(data)
df_no_cols.show()

df_with_cols = create_dataframe_basic_datatypes_with_column(data, columns)
df_with_cols.show()

df_with_schema = create_dataframe_with_schema(data_1, schema)
df_with_schema.show()
```

```
df_with_schema_1 = create_dataframe_with_schema(data_1, schema_1)
df_with_schema_1.show()

df_with_row_data = create_dataframe_with_schema(row_data, schema_1)
df_with_row_data.show()
```

```
df_complex_data = create_dataframe_with_schema(complex_type_data, schema_complex_type)
df_complex_data.show() 
```
