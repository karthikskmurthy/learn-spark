```
from pyspark.sql.functions import udf, col
from pyspark.sql.types import FloatType , StringType
```

```
dfFootballData = spark.read.option('inferSchema',True).option('header',True).csv('/FileStore/tables/CompleteDataset.csv')
display(dfFootballData)
```

```
import re
def convertToNumeric(input):
  numb = re.findall('\d+.*[KM]',input)
  res = 0
  if len(numb)>0:
    number = numb[0]
    if 'K' in number:
      res = float(number[0:len(number)-1])*(10**3)
    else:
      res = float(number[0:len(number)-1])*(10**6)
  return res
```

```
udf_func = udf(convertToNumeric,FloatType())
dfFootballCasted = dfFootballData.withColumn('wagecasted',udf_func(dfFootballData.Wage)).withColumn('valuecasted',udf_func(dfFootballData.Value))
```

```
dfFootballCasted.select('wage', 'value', 'wagecasted', 'valuecasted').show(5)
```
