```
rdd_logs = spark.sparkContext.textFile('/FileStore/tables/sample.log')
```

```
rdd_logs.take(5)
```

```
rdd_logs.map(lambda x: x.split(' ')).take(5)
```

```
import re
type_pattern = re.compile('\[\S*\]')
date_pattern = re.compile('(\d\d\d\d-\d\d-\d\d)')
```

```
rdd_log_type = rdd_logs.map(lambda x: type_pattern.findall(x)).map(lambda x: x[0] if len(x)>0 else '')
```

```
rdd_log_type.map(lambda x: (x, 1)).reduceByKey(lambda x,y:x+y).collect()
```

```
rdd_regex_parsed = rdd_logs.map(lambda x: (type_pattern.findall(x),date_pattern.findall(x), x ))
```

```
rdd_regex_parsed.collect()
```

```
rdd_parsed = rdd_regex_parsed.map(lambda x: ( x[0][0] if len(x[0])>0 else '',  x[1][0] if len(x[1])>0 else '', x[2]))
```

```
df_logs = rdd_parsed.toDF(['log_type', 'date', 'log'])
```

```
df_logs.show()
```
