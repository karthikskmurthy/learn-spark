## Filter and Where

```
from pyspark.sql.functions import col, when
df = spark.read.option('inferSchema', True).option('header', True).csv('/FileStore/tables/data.csv')
df.show(5)
```

```
df.where((col('open') > 12000) & (col('close') > 11500)).show(2)
df.where(col('open') > 12000).where(col('close') > 11500).show(2)
```

```
df.where((col('open') > 12000) | (col('close') > 12000)).show(2)
```

```
df.filter((col('open') < 11000) & (col('close') < 11000)).show(2)
```

## Sort and Order

```
df.sort(col('open')).show(2)
df.sort(col('open').desc(), col('close')).show(2)
```

```
df.orderBy(col('open')).show(2)
```

## Union

```
df1 = df.where(col('open') > 12000)
df2 = df.where(col('open') < 6000)
df3 = df1.union(df2)
df3.show(5)
```

## CASE

```
df.withColumn('candle', when(col('close') > col('open'), 'green').when(col('open') == col('close'), 'neutral').otherwise('red')).show(5)
```
