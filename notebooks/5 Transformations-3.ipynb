## Grouping

```
from pyspark.sql.functions import col, when, count, avg, to_date
from pyspark.sql.window import Window
df = spark.read.option('inferSchema', True).option('header', True).csv('/FileStore/tables/data.csv')
```

```
df.withColumn('candle', when(col('close') > col('open'), 'green').when(col('close') == col('open'), 'neutral').otherwise('red')) \
  .groupBy('candle') \
  .agg(count(col('Date')).alias('days')) \
  .orderBy(col('days')) \
  .show()
```

## Windowing

```
df_casted = df.withColumn('date_casted', to_date(col('Date'), 'dd-MMM-yyyy'))
df_ma = df_casted.withColumn('20DMA', avg(col('Close')) \
                               .over( Window.orderBy(col('date_casted')).rowsBetween(-20,Window.currentRow)) \
                     ).orderBy(col('date_casted').asc())
df_ma.orderBy(col('date_casted')).show()
```
