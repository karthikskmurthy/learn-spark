```
from pyspark.sql.functions import col,initcap, substring, trim, upper, lpad, instr, length, translate
```

```
df = spark.read.option('header', True).option('inferSchema', True).csv('/FileStore/tables/tripadvisor_hotel_reviews.csv')
df.show(5)
```

## INITCAP

```
df.withColumn('initial cap', initcap(col('review'))).show(2)
```

## SUBSTRING

```
df.withColumn('sub sring', substring(col('review'), 0, 4)).show(2)
```

## TRIM

```
df.withColumn('replace string', trim(col('review'))).show(2)
```

## UPPER

```
df.withColumn('upper', upper(col('review'))).show(2)
```

## PAD

```
df.withColumn('left pad', lpad(substring(col('review'), 0, 4), 10, '0')).show(2)
```

## INSTR

```
df.withColumn('in string', instr(col('review'), 'rooms')).show(2)
```

## LENGTH

```
df.withColumn('length', length(col('review'))).show(2)
```

## TRANSLATE

```
df.withColumn('translate', translate(col('review'), 'ne', '12')).show(5)
```
