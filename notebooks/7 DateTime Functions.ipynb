```
from pyspark.sql.functions import to_date, col, to_timestamp, unix_timestamp, from_unixtime, date_trunc, year, month, dayofmonth, dayofyear, current_date, current_timestamp, datediff
df = spark.read.option('inferSchema', True).option('header', True).csv('/FileStore/tables/data.csv')
df.show(2)
```

## to date

```
df.select('date', to_date(col('date'), 'dd-MMM-yyyy').alias('casted_date')).show(2)
```

## to timestamp

```
df.select('date', to_timestamp(col('date'), 'dd-MMM-yyyy').alias('casted_date')).show(2)
```

## unix timestamp

```
df.select('date', to_timestamp(col('date'), 'dd-MMM-yyyy').alias('casted_date')).select('*', unix_timestamp(col('casted_date')).alias('unix_timestamp')).show(2)
```

## from unix time

```
df.select('date', to_timestamp(col('date'), 'dd-MMM-yyyy').alias('casted_date')).select('*', unix_timestamp(col('casted_date')).alias('unix_timestamp')).select('*', from_unixtime(col('unix_timestamp')).alias('from_unixtime')).show(2)
```

## date trunc

```
df.select('date', to_timestamp(col('date'), 'dd-MMM-yyyy').alias('casted_date')).select('*', date_trunc('year', 'casted_date')).show(2)
```

## year month dayofmonth dayofyear

```
df.select('date', to_timestamp(col('date'), 'dd-MMM-yyyy').alias('casted_date')).select('*', year(col('casted_date')), month(col('casted_date')), dayofmonth(col('casted_date')), dayofyear(col('casted_date'))).show(2)
```

## current date

```
df.select('Date', current_date()).show(2)
```

## date diff

```
df.select('Date', 'date', to_date(col('date'), 'dd-MMM-yyyy').alias('casted_date'), current_date().alias('current_date')).select('*', datediff(col('current_date'), col('casted_date'))).show(2)
```
