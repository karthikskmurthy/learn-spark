```
from pyspark.sql.functions import col, max, min, avg, sum, count, countDistinct, approx_count_distinct, collect_list, first, last
```

```
df = spark.read.option('inferSchema', True).option('header', True).csv('/FileStore/tables/data.csv')
df.show(2)
```

## min max avg sum

```
df.select(max(col('high')), min(col('high')), avg(col('high')), sum(col('high'))).show()
```

## count countdistinct

```
df.select(count(col('high')), countDistinct(col('high')), approx_count_distinct(col('high'))).show()
```

## collect list

```
df.select(collect_list(col('high'))).show()
```

## first last

```
df.select(first(col('high')), last(col('high'))).show()
```
