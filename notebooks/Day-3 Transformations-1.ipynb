## Select and SelectExpr

```
from pyspark.sql.functions import col, expr, max, min, lit
df = spark.read.option('inferSchema', True).option('header', True).csv('/FileStore/tables/data.csv')
```

```
df.select('open', 'close').show(2)
df.select(col('open'), col('close')).show(2)
df.select(col('open'),'close').show(2)
```

```
df.select(expr('open')).show(2)
```

```
df.select('*', col('open').alias('alias_open')).show(2)
```

```
df.selectExpr('*', 'open as alias_open', '(open = low) as low_start').show(2)
```

```
df.agg(max('open').alias('max_open'), min('close').alias('min_close')).show()
```

```
df.selectExpr('max(open) as max_open', 'max(close) as max_close').show()
```

## With Column and With Column Renamed

```
df.withColumn('is_open_close', col('open') == col('close')).show(2)
```

```
df.withColumn('number', lit(1)).show(2)
```

```
df.withColumnRenamed('open', 'renamed_open').show(2)
```

## Drop

```
df.drop('open').show(2)
```

```
df.drop(col('open')).show(2)
```

## CAST

```
df.printSchema()
df.withColumn('casted_open', col('open').cast('string')).select('open', 'casted_open').show(2)
df.withColumn('casted_open', col('open').cast('string')).select('open', 'casted_open').printSchema()
```
